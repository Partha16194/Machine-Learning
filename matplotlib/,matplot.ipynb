```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
```

Create Line Plot

```python
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Line Plot')
plt.show()
```

Customized Plot

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,color='red',linestyle='--')
```

```python
plt.plot(x,y,color='red',linestyle='--',marker='o')
```

```python
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3)
```

```python
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
```

```python
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.grid(True)
```

Creating Subplots

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(x,y2,color='green')
plt.title("Plot 2")
```

Bar Plot

```python
categories=['A','B','C','D','E']
values=[5,7,3,8,6]

plt.bar(categories,values,color='purple')


```

Histograms

```python
data=[1,2,2,3,3,3,3,4,5,5,5]
plt.hist(data,bins=5,color='orange',edgecolor='black')
plt.grid(True)
```

Create Scatter Plot

```python
x=[1,2,3,4,5]
y=[2,4,7,9,5]

plt.scatter(x,y,color='blue',marker='x')
```

Creating Pie Chart

```python
labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightblue']
explode=[0.2,0,0,0]

plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True,explode=explode)

```

Analyzing Sales Data

```python
import pandas as pd 
data=pd.read_csv('Sales.csv',encoding = "ISO-8859-1")
data.head()


```

```python
#plot SALES  by DEALSIZE

sales_by_dealsize=data.groupby('DEALSIZE')['SALES'].sum()
sales_by_dealsize
```

```python
sales_by_dealsize.plot(kind='bar',color='purple')

#plt.grid(True)
```

```python
#Plotting Sales Over Time

sales_trend=data.groupby('ORDERDATE')['SALES'].sum().reset_index()
plt.plot(sales_trend['ORDERDATE'].head(),sales_trend['SALES'].head())
#sales_trend.plot(kind='line')
```
