Data Manipulation

```python
# Adding A New Column
data={
    'Name':['Krish','John','jack'],
    'Age':[25,30,35],
    'City':['Bangalore','New York','Florida']
    
}
df=pd.DataFrame(data)
df['Salary']=[50000,70000,90000]
df
```

```python
#Remove A Column

df.drop('Salary',axis=1)

#Here it is not a permanent operation
```

```python
#Dropping the column permanently
df.drop('Salary',axis=1,inplace=True)
df
```

```python
#Add age to the column 

df['Age']=df['Age']+1
df
```

```python
#Deleting A Particular Row Temporarily
df.drop(0)
```

```python
#Deleting It Permanently
df.drop(0,inplace=True)
df
```

Data Analysis

```python
df=pd.read_csv('Deepression.csv')
df.head(5)
```

```python
#Checking The Last 5 rows

df.tail(5)
```

```python
df.describe()
```

```python
# Handling Missing Values
```

```python
df.isnull()
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df_filled=df.fillna(0)
df_filled
```

```python
#Filling Missing Values With The Mean Of The Column

df['Sleep_FillNA']=df['Sleep'].fillna(df['Sleep'].mean())
df
```

```python
df.dtypes
```

```python
#Renaming Columns
df=df.rename(columns={'Sleep':'Sleeping'})
df.head()
```

```python
#Change Data Types

df['Sleep_FillNA_new']=df['Sleep_FillNA'].astype(int)
df.head()
```

```python
#Applying Function

df['New_Sleep']=df['Sleeping'].apply(lambda x:x*2)

df.head()
```

```python
#Data Aggregation & Grouping

grouped_mean=df.groupby('Depression State')['Interest'].mean()
grouped_mean
```

```python
grouped_sum=df.groupby('Depression State')['Appetite'].sum()
grouped_sum
```

```python
#Aggregate Multiple Functions

grouped_agg=df.groupby('Depression State')['Interest'].agg(['mean','sum','count'])
grouped_agg
```

```python
#Merging and Joining DataFrames
```

```python
df1=pd.DataFrame({'Key':['A','B','C'],'Value1':[1,2,3]})
df2=pd.DataFrame({'Key':['A','B','D'],'Value2':[4,5,6]})

```

```python
df1
```

```python
df2
```

```python
#Merging DataFrame On The Key Columns
```

```python
pd.merge(df1,df2,on="Key",how="inner")
```

```python
pd.merge(df1,df2,on="Key",how="outer")
```

```python
pd.merge(df1,df2,on="Key",how="left")
```

```python
pd.merge(df1,df2,on="Key",how="right")
```
